MODELO RED NEURONAL

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Definir las columnas que se van a normalizar
columnas_a_normalizar = ['Edad', 'Ingreso', 'Ofertas_Laborales', 'Experiencia_Laboral_Previa', 'Satisfaccion', 'Serie']

# Crear el objeto MinMaxScaler para escalar los datos entre 0 y 1
scaler = MinMaxScaler()

# Normalizar los datos en el rango [0, 1]
data_normalizada = scaler.fit_transform(dataframe_respuestas[columnas_a_normalizar])

# Convertir los datos normalizados en un DataFrame
data_normalizada = pd.DataFrame(data_normalizada, columns=columnas_a_normalizar)

# Agregar la columna 'Continuidad' de vuelta
data_normalizada['Continuidad'] = dataframe_respuestas['Continuidad']

# Definir los pesos
pesos = {
    'Edad': 0.2,
    'Ingreso': 0.5,
    'Ofertas_Laborales': 0.1,
    'Experiencia_Laboral_Previa': 0.1,
    'Satisfaccion': 0.05,
    'Serie': 0.05
}

# Crear una nueva columna 'Peso' en el DataFrame normalizado basada en las características y sus pesos
data_normalizada['Peso'] = (data_normalizada['Edad'] * pesos['Edad'] +
                            data_normalizada['Ingreso'] * pesos['Ingreso'] +
                            data_normalizada['Ofertas_Laborales'] * pesos['Ofertas_Laborales'] +
                            data_normalizada['Experiencia_Laboral_Previa'] * pesos['Experiencia_Laboral_Previa'] +
                            data_normalizada['Satisfaccion'] * pesos['Satisfaccion'] +
                            data_normalizada['Serie'] * pesos['Serie'])

# Mostrar el DataFrame con los datos normalizados y la nueva columna de pesos
print(data_normalizada.head())

```

Crear de la red neuronal

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Definir la función de activación (Escalón binario)
def activacion(datos, pesos, bias):
    suma = np.dot(datos, pesos) + bias  # Cálculo de la suma
    return 1 if suma >= 0 else 0

# Función para recibir los datos de entrada
def data_resivida():
    print("Por favor, ingrese los siguientes datos:")
    edad = float(input("Edad: "))
    ingreso = float(input("Ingreso: "))
    ofertas_laborales = float(input("Ofertas Laborales: "))
    experiencia_laboral_previa = float(input("Experiencia Laboral Previa: "))
    satisfaccion = float(input("Satisfacción: "))
    serie = float(input("Serie: "))

    # Crear DataFrame para los datos de entrada
    datos_entrada = pd.DataFrame([[edad, ingreso, ofertas_laborales, experiencia_laboral_previa, satisfaccion, serie]],
                                  columns=['Edad', 'Ingreso', 'Ofertas_Laborales', 'Experiencia_Laboral_Previa', 'Satisfaccion', 'Serie'])

    # Estandarizar los datos de entrada con el mismo scaler que fue usado antes
    datos_entrada[['Edad', 'Ingreso', 'Ofertas_Laborales', 'Experiencia_Laboral_Previa', 'Satisfaccion', 'Serie']] = scaler.transform(datos_entrada[['Edad', 'Ingreso', 'Ofertas_Laborales', 'Experiencia_Laboral_Previa', 'Satisfaccion', 'Serie']])

    return datos_entrada.values.flatten()


# Crear el objeto MinMaxScaler y ajustar a los datos normalizados
scaler = MinMaxScaler()
scaler.fit(data_normalizada[['Edad', 'Ingreso', 'Ofertas_Laborales', 'Experiencia_Laboral_Previa', 'Satisfaccion', 'Serie']])

# Características y clase objetivo
X = data_normalizada[['Edad', 'Ingreso', 'Ofertas_Laborales', 'Experiencia_Laboral_Previa', 'Satisfaccion', 'Serie']].values
y = data_normalizada['Continuidad'].values

# Inicializar pesos aleatoriamente y bias
np.random.seed(0)  # Para reproducibilidad
pesos = np.random.rand(X.shape[1])  # Inicializa un peso aleatorio por cada característica
bias = np.random.rand()  # Inicializa el bias aleatoriamente
tasa_de_aprendizaje = 0.01  # Tasa de aprendizaje
epocas = 100  # Número de épocas de entrenamiento

# Entrenamiento del perceptrón
for epoca in range(epocas):
    error_total = 0
    for i in range(len(X)):
        datos = X[i]  # Características de la fila i
        prediccion = activacion(datos, pesos, bias)  # Predicción inicial
        error = y[i] - prediccion  # Cálculo del error
        error_total += error**2  # Acumulación del error cuadrático

        # Actualización de pesos y bias
        pesos += tasa_de_aprendizaje * error * datos
        bias += tasa_de_aprendizaje * error

        # Imprimir para depuración
        print(f"Datos: {datos}, Predicción: {prediccion}, Error: {error}, Pesos: {pesos}, Bias: {bias}")

    # Monitoreo del progreso en cada época
    print(f"Época {epoca + 1}, Error total: {error_total}")

# Función para predecir con nuevos datos
def predecir(nuevos_datos):
    prediccion_final = activacion(nuevos_datos, pesos, bias)  # Usar la función de activación
    return prediccion_final

#uso de nuevos datos
nuevos_datos = data_resivida()  # Obtener nuevos datos del usuario


```

**RESULTADOS DEL APRENDIZAJE DEL MODELO [ PREDICCION DE CONTINUIDAD ]**

```python
resultado = predecir(nuevos_datos)  # Pasar todos los datos
if resultado==1:
  print(f"Predicción de continuidad:[ {resultado} ]", "El estudiante continuará con la carrera universitaria")

else:
  print(f"Predicción de continuidad: [{resultado} ]", "El estudiante  NO continuará con la carrera universitaria")
```
